# Palindromes

```python
import random
import string

def generate_palindromes(n, max_length=10):
    """ Generate 'n' palindromes with a maximum length of 'max_length'. """
    palindromes = set()

    while len(palindromes) < n:
        length = random.randint(1, max_length // 2)
        half_palindrome = ''.join(random.choices(string.ascii_lowercase, k=length))
        palindrome = half_palindrome + half_palindrome[::-1]
        palindromes.add(palindrome)

    return list(palindromes)

# Generating 50 lowercase alphabetical palindromes
palindrome_list = generate_palindromes(50)

def generate_non_palindromes(n, max_length=10):
    """ Generate 'n' non-palindromes with a maximum length of 'max_length'. """
    non_palindromes = set()

    while len(non_palindromes) < n:
        length = random.randint(1, max_length)
        s = ''.join(random.choices(string.ascii_lowercase, k=length))
        # Ensure that the string is not a palindrome
        if s != s[::-1]:
            non_palindromes.add(s)

    return list(non_palindromes)

# Generating 50 lowercase alphabetical non-palindromes
non_palindrome_list = generate_non_palindromes(50)
data = palindrome_list + non_palindrome_list
targets = [i<len(palindrome_list) for i in range(len(data))]
```
